```python
import pandas as pd
import os

input_path = "/group/moniergrp/dbaral/run_project/output_data/historical_model"
output_path = "/group/moniergrp/dbaral/run_project/output_data/historical_model"

df_metrics = pd.read_csv(os.path.join(input_path,"final_metrics.csv"))

# define bad iterations (example: negative R²)
bad_iterations = df_metrics[df_metrics["R2_test"] <= 0.2]["iteration"].unique()

print("Bad iterations:", bad_iterations)

#filter
df_clean_metrics = df_metrics[df_metrics["R2_test"] > 0.2]
#svae
df_clean_metrics.to_csv(os.path.join(output_path, "final_cleaned_metrics.csv"), index=False)
#now coefficients file
df_coef = pd.read_csv(os.path.join(input_path, "final_coefficients.csv")) 

# REMOVE bad iterations
df_coef_clean = df_coef[~df_coef["iteration"].isin(bad_iterations)]

# save
df_coef_clean.to_csv(os.path.join(output_path, "final_cleaned_coefficients.csv"), index=False)

print("Saved cleaned coefficients file")
```
